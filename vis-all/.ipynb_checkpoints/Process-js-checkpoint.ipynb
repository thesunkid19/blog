```python
import json
import os
```

```python
img_dir = '/media/th/hdd1/dataset/coco-img/val2017'
json_f = './vis.json'
```

```python
js = json.load(open(json_f))
img = os.listdir(img_dir)
```

```python
itsect = [el for el in js if str("%012d"%int(el['image_id'])+'.jpg') in img]
```

```python
itsect = [{'image_id':str("%012d"%int(i['image_id'])), 'caption':i['caption']} for i in itsect]
```

```python
itsect
```

### Groundtruth file

```python
gt_dir = '../data/dataset_coco.json'
gt = json.load(open(gt_dir))
```

```python
print(gt['images'][0]['sentences'][0]['raw'])
print(gt['images'][0]['filename'][13:-4])
```

```python
name_to_id = {}
for i,obj in enumerate(gt['images']):
    name_to_id[obj['filename'][13:-4]] = i
```

```python
for el in itsect:
    _id = name_to_id[el['image_id']]
    assert(len(gt['images'][_id]['sentences'])==5, "no cap per img != 5")
```

### SGAE

```python
sg_dir = '../vis-sgae/_vis.json'
sg = json.load(open(sg_dir))
```

```python
name_to_idsg = {}
for i,obj in enumerate(sg):
    name_to_idsg[obj['image_id']] = i
```

```python
for el in itsect:
    _id = name_to_id[el['image_id']]
    _id2 = name_to_idsg[el['image_id']]
    el['sgcaption'] = sg[_id2]['caption'] # sgae to js
    for i in range(5):
        el['gtcaption'+str(i+1)] = gt['images'][_id]['sentences'][i]['raw'] #gt to json
```

### Save json & move img

```python
json.dump(itsect, open('_vis.json', 'w'))
```

```python
itsect_img = [i['image_id']+'.jpg' for i in itsect]
```
